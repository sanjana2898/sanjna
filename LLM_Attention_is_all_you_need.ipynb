```python

'''

Stage 1 → Dataset download + preprocessing
Stage 2 → Load GPT2, modify for classification
Stage 3 → Fine-tune and evaluate

##[text batch] → GPT-2 → predictions → compare to correct label → calculate loss → update weights

'''

import pandas as pd
import zipfile
import requests
import io
import torch.nn as nn


#load pretrained model

from transformers import GPT2LMHeadModel, GPT2Tokenizer,GPT2Config,GPT2Model
from torch.utils.data import Dataset, DataLoader
import torch

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np


model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# 🔹 Step 1: Download the ZIP file from UCI
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00228/smsspamcollection.zip"
response = requests.get(url)

# 🔹 Step 2: Extract the SMSSpamCollection file from ZIP
with zipfile.ZipFile(io.BytesIO(response.content)) as z:
    with z.open("SMSSpamCollection") as f:
        # 🔹 Step 3: Load into pandas
        df = pd.read_csv(f, sep='\t', header=None, names=["label", "text"]) #✅ Now df has 2 columns: label (ham/spam) and text (SMS message).



# 🔹 Step 4: Save as CSV locally
df.to_csv("sms_spam_collection.csv", index=False)

# 🔹 Step 5: Show first few rows
print(df.head())


##################################
#Preprocess the dataset for actual model input


#We convert label (ham/spam) to numbers:

#ham → 0

#spam → 1
#######################################



# Convert labels to numeric for training


df["label"] = df["label"].map({"ham": 0, "spam": 1})


#For training with PyTorch, we need a custom Dataset and DataLoader. -- Create batches (groups) of data to feed the model

'''
Dataset & DataLoader: These help us organize and load data in batches. Think of them like waiters that serve data to the model in chunks.
 Create a Custom Dataset Class
 We define a custom dataset class (so PyTorch knows how to read and prepare our SMS data).

def __init__(self, dataframe, tokenizer, max_length=128):
    self.data = dataframe          # Our SMS data
    self.tokenizer = tokenizer     # GPT-2 tokenizer
    self.max_length = max_length   # Maximum text length

Tells PyTorch how many SMS messages we have.
def __len__(self):
    return len(self.data)

So if we have 5000 SMS messages, this returns 5000.


🔹 __getitem__: What happens when you ask for SMS #10, or #25?
def __getitem__(self, idx):
    text = self.data.iloc[idx]["text"]    # The SMS text
    label = self.data.iloc[idx]["label"]  # 0 (ham) or 1 (spam)

Then we tokenize the text using the GPT-2 tokenizer:

    enc = self.tokenizer(text,
                         truncation=True,         # Cut off long text
                         padding="max_length",    # Make it fixed length
                         max_length=self.max_length,
                         return_tensors="pt")     # Return PyTorch tensors

Finally, we return the processed data:

    return {
        "input_ids": enc["input_ids"].squeeze(),          # The tokenized text as numbers
        "attention_mask": enc["attention_mask"].squeeze(),# 1 for real token, 0 for padded
        "labels": torch.tensor(label, dtype=torch.long)   # The label as a number
    }

dataset: Creates an object of all our SMS messages prepared for training.

DataLoader: Splits the dataset into small batches of 4 messages.

shuffle=True: Mixes up the messages every epoch to prevent overfitting.

It is returning one item (one message) in a special format that a model like GPT-2 expects.

attention mask - It tells the model which tokens are real words (1) and which are just padding (0).

"labels": torch.tensor(label, dtype=torch.long)
This converts the label (e.g., "spam" or "ham") into:

0 for ham (not spam)

1 for spam
And wraps it as a PyTorch tensor.📌 So this tells the model: “this message is spam (1)” or “not spam (0)”.


{
  "input_ids": tensor([...]),         # Numbers that represent the message
  "attention_mask": tensor([...]),    # Mask that says which parts are real
  "labels": tensor(0 or 1)            # The correct answer (spam or not)
}

so therefore -


 return {
            "input_ids": enc["input_ids"].squeeze(),  # tensor of token ids
            "attention_mask": enc["attention_mask"].squeeze(),  # which tokens are real vs padded
            "labels": torch.tensor(label, dtype=torch.long)  # 0 or 1
        }

dataset = SMSDataset(df, tokenizer)
loader = DataLoader(dataset, batch_size=4, shuffle=True)


addons -

config = GPT2Config.from_pretrained("gpt2")??
Loads the configuration settings for GPT-2.

Think of it as the blueprint of the model: how many layers it has, hidden size, vocab size, etc.

next -

gpt2_base = GPT2Model.from_pretrained("gpt2", config=config)
What this does:

Actually loads the GPT-2 model weights from the internet.

We tell it: “Use the config we just loaded.”

This loads only the base model — it does NOT include the language head (i.e., the final layer used to predict next words or classify).

Think of it as:
“Give me the brain of GPT-2 without the mouth (prediction layer).”

next--


 gpt2_base.resize_token_embeddings(len(tokenizer))->“Hey GPT-2, your vocabulary just got a little bigger. Please update yourself!”

 GPT-2 was trained with 50257 tokens, but we just added one more (<pad> token) earlier:

tokenizer.pad_token = tokenizer.eos_token
So we need to increase GPT-2's vocabulary size to match the tokenizer.

📏 Why? Because otherwise the model won’t know how to deal with this new padding token.

Line	What it Means
GPT2Config...	Load the blueprint for GPT-2 (settings) 🧾
GPT2Model.from_pretrained...	Load the actual brain (model) 🧠
.resize_token_embeddings...	Tell GPT-2: "You now have one more word in your dictionary!" 🧠➕🧾


'''



from torch.utils.data import Dataset, DataLoader
from transformers import GPT2Tokenizer
import torch

# Load GPT-2 tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token  # Add pad token

# Define custom dataset class
class SMSDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_length=128):
        self.data = dataframe
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        text = self.data.iloc[idx]["text"]
        label = self.data.iloc[idx]["label"]

        # Tokenize the text
        enc = self.tokenizer(text,
                             truncation=True,
                             padding="max_length",
                             max_length=self.max_length,
                             return_tensors="pt")

        return {
            "input_ids": enc["input_ids"].squeeze(),  # tensor of token ids
            "attention_mask": enc["attention_mask"].squeeze(),  # which tokens are real vs padded
            "labels": torch.tensor(label, dtype=torch.long)  # 0 or 1
        }

# Create dataset and dataloader
dataset = SMSDataset(df, tokenizer)
loader = DataLoader(dataset, batch_size=4, shuffle=True)

# Load GPT2 base model
#task - We’re loading the base GPT-2 model (just the "brain" of GPT-2 without the final layer that makes predictions
config = GPT2Config.from_pretrained("gpt2")  #
gpt2_base = GPT2Model.from_pretrained("gpt2", config=config) #
gpt2_base.resize_token_embeddings(len(tokenizer))  #
#

#Modify the model for classification

# We add a Linear layer to classify spam/ham.


'''
note-
GPT-2 is a language model — it takes some text and tries to predict the next word.

But in our case, we want it to predict a label:

🔸 "ham" → 0
🔸 "spam" → 1

So we need to change the end of GPT-2.

What is Step 6 doing?
We are adding a final "classifier" layer (just a small neural network block at the top) like this:

            ┌──────────────────────────────┐
            │      GPT-2 Model             │
            │  (returns many hidden states)│
            └────────────┬─────────────────┘
                         │
      (just take last token's vector)
                         ↓
        [ ... 768 numbers ... ] ←  vector of last token
                         ↓
              🧠 Linear Layer (size 768 → 2)
                         ↓
                 [score_for_ham, score_for_spam]
                         ↓
              Final answer = highest score


'''

'''
 We're creating a new class (a blueprint) that builds a custom GPT-2 model for classifying text (e.g., spam or not spam). It’s based on PyTorch's neural network module (nn.Module).


'''

# We add a Linear layer to classify spam/ham. for my explicity add on of classigfication logic

class GPT2ForClassification(nn.Module):  # we are making our own GPT2 classifier
    def __init__(self, base_model, n_classes=2):
        super().__init__()
        self.gpt2 = base_model                          # Base GPT-2 model -- We're saving the original GPT-2 model so we can use it inside our custom classifier.
        self.dropout = nn.Dropout(0.3)                  # Regularization (to avoid overfitting)-- Dropout randomly turns off 30% of the neurons during training to prevent the model from memorizing the training data (helps generalize better).
        self.classifier = nn.Linear(768, n_classes)     # Our classifier (768 → 2) -- This is a small layer that takes a vector of size 768 (GPT-2’s output size) and turns it into class scores (like spam vs. ham).

    def forward(self, input_ids, attention_mask, labels=None):
        # 🚀 Pass input into GPT-2
        outputs = self.gpt2(input_ids=input_ids, attention_mask=attention_mask) ## -- We send the input text into GPT-2 and get its output. The attention_mask tells GPT-2 which words to pay attention to.



        # 📦 Take the hidden states from GPT-2
        last_hidden = outputs.last_hidden_state         # shape: [batch, seq_len, 768] -- We grab the hidden states (fancy word for GPT-2’s internal understanding) of all the tokens (words) in the sentence.

        # 🧠 Take the last token’s vector from the sentence
        pooled = last_hidden[:, -1, :]                  # shape: [batch, 768]   We pick just the vector from the last word in each sentence. This is often used as a summary of the whole sentence.

        # 🎯 Pass through classifier to get spam/ham prediction scores
        logits = self.classifier(self.dropout(pooled))  # shape: [batch, 2] We apply dropout, then feed the result into our classifier. The output (logits) is a score for each class (e.g., how likely it's spam or ham).



        if labels is not None: #If we also have the correct labels (like during training)...
            # 📉 Calculate loss if labels are provided (during training)
            loss_fn = nn.CrossEntropyLoss()   #We use a standard loss function that measures how far off our predictions are from the correct labels.
            loss = loss_fn(logits, labels) # We compute the loss — a number that tells us how bad our prediction was.
            return loss, logits #e return both the loss and the predicted scores, which are used to update the model.

        return logits  # 🧾 During prediction, return just the class scores




#####

'''
You are creating a new model called model — but not just any model.

You are creating an enhanced version of GPT-2 by:

Wrapping the base GPT-2 model (gpt2_base) inside your custom classification class GPT2ForClassification.

This new model:

Uses GPT-2 to convert text into meaningful embeddings.

Then uses a linear classifier (a small neural layer) to predict whether a message is spam or ham (binary class).
Think of gpt2_base like a brain that understands English sentences.

But it doesn't know how to say “This is spam” or “This is ham”.

So GPT2ForClassification adds a mouth — a classifier — that takes GPT-2's understanding and speaks a decision:

"Spam" (1) or "Ham" (0)


model = GPT2ForClassification(gpt2_base) -- “Build me a spam/ham classifier, powered by GPT-2's language understanding.”

device in PyTorch is just a pointer that tells your model or data:
"Hey, should I run this on the CPU or GPU?"

It doesn’t do any computation itself.
It’s like a shipping label you stick on the model or tensor saying:

“Send this to GPU”

or “Keep this on CPU”

model.to(device) --
 This moves your entire model (its layers, weights, and memory) to the selected device.

Why? Because during training, both the model and data must be on the same device (GPU or CPU).
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)
his defines how the model learns — called the optimizer.

AdamW: A smart optimizer that adapts learning for each weight. It's like an improved version of classic SGD.

model.parameters(): These are the model weights we want to update during training.

lr=2e-5: This is the learning rate → it controls how big the steps are when updating weights (small = safer learning).

Line	Purpose
torch.device(...)	Choose GPU if available
model.to(device)	Move model to that device
torch.optim.AdamW(...)	Choose optimizer for learning weights



An optimizer decides:

“How should I change the model’s weights so it makes fewer mistakes?”

What is AdamW?
➤ Adam = Adaptive Moment Estimation
It’s smart. It learns how fast and in what direction to update each weight.

➤ AdamW = Adam + weight decay
It also penalizes overly large weights → avoids overfitting.

optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)

📦 Track gradients of all weights in model.parameters()

⚖️ Use adaptive learning rate for each parameter (Adam learns this from past gradient trends)

🧮 Perform this update:

weight = weight - adjusted_learning_rate * gradient - weight_decay_term
Why is AdamW Better than Adam?
Adam can sometimes ignore regularization.

AdamW fixes this by cleanly decaying weights after every update — helps generalization.

Concept	Meaning
Optimizer	Learns how to update weights to reduce loss
Gradient	Direction of error change
Adam	Uses momentum + adaptive learning
Weight Decay (W)	Keeps weights small to avoid overfitting
AdamW	Best of both worlds: smart + regularized


📢 Put the model in training mode. This activates layers like dropout which are only used while learning.



model.train()
 Loop over your data 2 full times (called epochs). More epochs = more learning.

 for epoch in range(2):  # train for 2 epochs
📊 Keep track of how much total error (loss) we accumulate.

    total_loss = 0
    for batch in loader:
📦 Go through the data in mini-batches (4 examples at a time, as set earlier). This is more efficient than doing one by one.

        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

📤 Send the data (tokenized text and labels) to the same device as the model (GPU/CPU).

        optimizer.zero_grad()
Every time we calculate loss, gradients are stored. If we don't clear them, they pile up.

        loss, logits = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
 Pass the data to the model.

logits = model's raw predictions (before softmax)

loss = how far those predictions are from the correct label

        loss.backward()
🔍 Do backpropagation — calculate how much each weight in the model contributed to the error.

        optimizer.step()
 Update the model's weights using AdamW, so it performs a little better next time.

         total_loss += loss.item()
Add this batch's loss to the running total for the epoch.

    print(f"Epoch {epoch+1} Loss: {total_loss:.4f}")
🖨️ Print the total loss after finishing one round over the dataset. Lower = better.

“Spam or not?”

The model guesses.

The teacher corrects it (loss).

The model writes notes (gradients).

The model updates its notes (optimizer.step()).

Repeat this for all SMS messages → that's 1 epoch.

--




'''

# Initialize classification model
model = GPT2ForClassification(gpt2_base)



#Optimizer and move to GPU/CPU if available.


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)

#Finetune the model -- We train for a few epochs.

#this is done because - train the model now, not evaluate it.
'''
"train the model now, not evaluate it. Turn ON dropout, etc."


'''
model.train()

'''
You're going to show the entire dataset 2 times to the model.
Each round is called an epoch

for epoch in range(2):  # train for 2 epochs
    total_loss = 0  #You’ll track how much "mistake" (loss) the model made in this epoch.
    #At the end, you’ll print how well the model did.


    for batch in loader:
 You split your data into small groups (called batches).
This goes through each batch one by one.


Inside the batch loop

        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

You get the input (text), mask (which parts are real), and label (spam or not).

.to(device) moves data to the GPU if you have one (faster!).


        optimizer.zero_grad() - “Start fresh. Forget what you learned last step.”
        loss, logits = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)

ou run the model on your batch.

logits: The model's guess for each message (like: "I think this is 0.8 spam").

loss: The model's mistake — how far its guess was from the truth.

        loss.backward()

“Based on the mistake you made, update your brain!”
This is called backpropagation — the learning part.

        optimizer.step()
 This updates the model's weights (like its brain cells) to do better next time.

         total_loss += loss.item()

 Add the current mistake (loss) to the total — so at the end of this epoch, you know how much total error the model made.

    print(f"Epoch {epoch+1} Loss: {total_loss:.4f}")
Print the total loss (error) after 1 full pass through the data.
If this number goes down each time — your model is learning!

Real Life Analogy
You're training a student on SMS messages. Each loop is like:

Give them 4 messages (a batch)

They guess: spam or not spam

You tell them if they're wrong (loss)

They fix their understanding (backprop)

Repeat this for all messages (1 epoch)

Then do it again (2nd epoch) — now they're smarter





'''



for epoch in range(2):  # train for 2 epochs
    total_loss = 0
    for batch in loader:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        optimizer.zero_grad()
        loss, logits = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1} Loss: {total_loss:.4f}")

#Evaluate Fine-tuned Model (Accuracy, Precision, Recall, F1)
'''
We’ll test the model on the same dataset for now (for simplicity), using evaluation metrics:

Accuracy → Overall how many did it get right?

Precision → Of the ones it said were spam, how many were actually spam?

Recall → Of all real spam, how many did it catch?

F1 Score → Balance between Precision & Recall



'''
#TESTING -

# Put model in evaluation mode (turns off dropout etc.)
model.eval()

# Store all predictions and labels
all_preds = []
all_labels = []

# No need to track gradients while evaluating
with torch.no_grad():
    for batch in loader:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        logits = model(input_ids=input_ids, attention_mask=attention_mask)  # get only logits, not loss

        preds = torch.argmax(logits, dim=1)  # get predicted class (0 or 1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

# 📊 Calculate metrics
acc = accuracy_score(all_labels, all_preds)
prec = precision_score(all_labels, all_preds)
rec = recall_score(all_labels, all_preds)
f1 = f1_score(all_labels, all_preds)

# 🖨️ Show results
print("✅ Evaluation Metrics:")
print(f"🔹 Accuracy:  {acc:.4f}")
print(f"🔹 Precision: {prec:.4f}")
print(f"🔹 Recall:    {rec:.4f}")
print(f"🔹 F1 Score:  {f1:.4f}")



###############################


#Use Model on New SMS Message

def predict_spam(text):
    model.eval()

    # Tokenize new text
    enc = tokenizer(text, return_tensors='pt', truncation=True, padding='max_length', max_length=128)
    input_ids = enc['input_ids'].to(device)
    attention_mask = enc['attention_mask'].to(device)

    # Predict
    with torch.no_grad():
        logits = model(input_ids=input_ids, attention_mask=attention_mask)
        pred = torch.argmax(logits, dim=1).item()

    label = "Spam" if pred == 1 else "Ham"
    return label

## testing
print(predict_spam("Win a free vacation now!!! Call 12345678"))
print(predict_spam("Are we still meeting at 5pm?"))






'''
Step	Description	Status ✅
1–3	Download, preprocess, tokenize	✅ Done
4–6	Load GPT2, modify for classification	✅ Done
7	Optimizer and training utilities	✅ Done
8	Fine-tune model	✅ Done
9	Evaluate model (accuracy, etc.)	✅ Done Now
10	Predict on new messages	✅ Live Function


'''


```

```python
#Define GPT2 Classifier Model

config = GPT2Config.from_pretrained("gpt2")
gpt2_base = GPT2Model.from_pretrained("gpt2", config=config)
gpt2_base.resize_token_embeddings(len(tokenizer))

class GPT2ForClassification(nn.Module):
    def __init__(self, base_model, n_classes=2):
        super().__init__()
        self.gpt2 = base_model
        self.dropout = nn.Dropout(0.3)
        self.classifier = nn.Linear(768, n_classes)

    def forward(self, input_ids, attention_mask, labels=None):
        outputs = self.gpt2(input_ids=input_ids, attention_mask=attention_mask)
        last_hidden = outputs.last_hidden_state
        pooled = last_hidden[:, -1, :]
        logits = self.classifier(self.dropout(pooled))
        if labels is not None:
            loss_fn = nn.CrossEntropyLoss()
            loss = loss_fn(logits, labels)
            return loss, logits
        return logits

model = GPT2ForClassification(gpt2_base)
```

```python
#Training Setup

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)
```

```python
import time

model.train()
for epoch in range(2):  # 2 epochs
    total_loss = 0
    print(f"\n🔁 Starting Epoch {epoch+1}")
    batch_num = 0

    for batch in loader:
        batch_num += 1
        start_time = time.time()  # Track how long each batch takes

        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        print(f"\n🧾 Batch {batch_num}:")
        print(f"  🔹 input_ids shape: {input_ids.shape}")
        print(f"  🔹 attention_mask shape: {attention_mask.shape}")
        print(f"  🔹 labels: {labels}")

        optimizer.zero_grad()
        loss, _ = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        print(f"  🔸 Loss: {loss.item():.4f}")

        loss.backward()
        optimizer.step()

        end_time = time.time()
        batch_time = end_time - start_time
        print(f"  ⏱️ Batch time: {batch_time:.2f}s")

        total_loss += loss.item()

    print(f"\n✅ Epoch {epoch+1} Total Loss: {total_loss:.4f}")
```

```python
'''
Overview of the Fine-Tuning Process
Data Preparation: Load and preprocess the dataset.

Tokenization: Convert text into tokens suitable for model input.

Model Definition: Define a simple language model architecture.

Training: Train the model on the prepared dataset.

Evaluation: Assess the model's performance on unseen data.


Stage 1


Data Preparation
'''
'''
We read the text file and convert it to lowercase to reduce the vocabulary size.

Non-alphabetic characters are removed to simplify the dataset.

The text is split into individual words for tokenization.


'''

import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import torch.nn.functional as F
import random
import re

# Set random seed for reproducibility
torch.manual_seed(42)

# Load and preprocess the dataset
with open('the-verdict.txt', 'r', encoding='utf-8') as f:
    text = f.read()

# Basic preprocessing: lowercase and remove non-alphabetic characters
text = text.lower()
text = re.sub(r'[^a-z\s]', '', text)
words = text.split()


'''
 Tokenization
We'll create a simple tokenizer that maps each unique word to an integer.
'''

'''
Explanation:

word_to_ix maps each word to a unique index.

ix_to_word allows us to convert indices back to words.


'''

# Create a vocabulary of unique words
vocab = sorted(set(words))
word_to_ix = {word: i for i, word in enumerate(vocab)}
ix_to_word = {i: word for i, word in enumerate(vocab)}




'''
 Dataset and DataLoader
We'll define a custom dataset class to generate input-target pairs for training.
'''

class TextDataset(Dataset):
    def __init__(self, words, word_to_ix, context_size=4):
        self.data = []
        for i in range(len(words) - context_size):
            context = words[i:i+context_size]
            target = words[i+context_size]
            self.data.append((context, target))
        self.word_to_ix = word_to_ix

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        context, target = self.data[idx]
        context_idxs = torch.tensor([self.word_to_ix[w] for w in context], dtype=torch.long)
        target_idx = torch.tensor(self.word_to_ix[target], dtype=torch.long)
        return context_idxs, target_idx





'''

Explanation:

For each position in the text, we take the previous context_size words as input and the next word as the target.

The dataset returns pairs of input indices and the target index.

The DataLoader batches the data and shuffles it for training.

'''
# Create dataset and dataloader
context_size = 4
dataset = TextDataset(words, word_to_ix, context_size)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)

'''
Model Definition
We'll define a simple neural network for language modeling.
The model consists of an embedding layer, a hidden linear layer with ReLU activation, and an output layer that predicts the next word.

Inputs are sequences of word indices, which are converted to embeddings and then passed through the network.
'''


class LanguageModel(nn.Module):
    def __init__(self, vocab_size, embedding_dim, context_size):
        super(LanguageModel, self).__init__()
        self.embeddings = nn.Embedding(vocab_size, embedding_dim)
        self.linear1 = nn.Linear(context_size * embedding_dim, 128)
        self.linear2 = nn.Linear(128, vocab_size)

    def forward(self, inputs):
        embeds = self.embeddings(inputs).view((inputs.size(0), -1))
        out = F.relu(self.linear1(embeds))
        out = self.linear2(out)
        return out
# Training the Model
#We'll train the model using cross-entropy loss and the Adam optimizer.

'''
Explanation:

We train the model for a specified number of epochs.

For each batch, we compute the loss, perform backpropagation, and update the model parameters.

The loss is printed after each epoch to monitor training progress.
'''

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Initialize model, loss function, and optimizer
embedding_dim = 64
model = LanguageModel(len(vocab), embedding_dim, context_size).to(device)
loss_function = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Training loop
epochs = 10
for epoch in range(epochs):
    total_loss = 0
    for context_idxs, target_idx in dataloader:
        context_idxs = context_idxs.to(device)
        target_idx = target_idx.to(device)

        model.zero_grad()
        logits = model(context_idxs)
        loss = loss_function(logits, target_idx)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")

#Evaluation
#We'll evaluate the model by generating text based on a given context.

'''
Explanation:

Starting from a given context, we generate the next word repeatedly to produce a sequence of text.

The model predicts the next word based on the current context.

The generated text is printed for inspection.


'''

def generate_text(model, start_context, length):
    model.eval()
    context = start_context.lower().split()
    generated = context.copy()
    for _ in range(length):
        context_idxs = torch.tensor([word_to_ix.get(w, 0) for w in context[-context_size:]], dtype=torch.long).unsqueeze(0).to(device)
        with torch.no_grad():
            output = model(context_idxs)
            predicted_idx = torch.argmax(output, dim=1).item()
            predicted_word = ix_to_word[predicted_idx]
            generated.append(predicted_word)
    return ' '.join(generated)

# Example usage
start_context = "the court finds"
generated_text = generate_text(model, start_context, 20)
print(generated_text)
```

```python
'''
Here is your fully updated end-to-end script with:

 Proper multi-head attention masking
All stages (data loading, model, training, generation, evaluation)
 GPT-style Transformer with dropout, feedforward, layer norm, causal attention
 Working text generation and perplexity
 Sample test predictions

 '''


import torch, re, requests, math
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

# 🔒 Set seed for reproducibility
torch.manual_seed(42)

# ✅ Stage 1: Load and preprocess text
url = "https://raw.githubusercontent.com/rasbt/LLMs-from-scratch/main/ch02/01_main-chapter-code/the-verdict.txt"
text = requests.get(url).text.lower()
text = re.sub(r'[^a-z\s]', '', text)
words = text.split()

# ✅ Stage 2: Tokenization
vocab = sorted(set(words))
word_to_ix = {w: i for i, w in enumerate(vocab)}
ix_to_word = {i: w for i, w in enumerate(vocab)}

# ✅ Stage 3: Dataset and DataLoader
class TextDataset(Dataset):
    def __init__(self, words, word_to_ix, context_size=4):
        self.data = [(words[i:i+context_size], words[i+context_size])
                     for i in range(len(words) - context_size)]
        self.word_to_ix = word_to_ix

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        ctx, tgt = self.data[idx]
        ctx_ids = torch.tensor([self.word_to_ix[w] for w in ctx])
        tgt_id = torch.tensor(self.word_to_ix[tgt])
        return ctx_ids, tgt_id

context_size = 4
dataset = TextDataset(words, word_to_ix, context_size)
loader = DataLoader(dataset, batch_size=64, shuffle=True)

# ✅ Stage 4: GPT-style Transformer Model with Masking
def generate_causal_mask(B, T, device, heads):
    base_mask = torch.tril(torch.ones(T, T, device=device)) == 0  # shape: [T, T]
    return base_mask.unsqueeze(0).expand(B * heads, T, T)  # shape: [B*heads, T, T]

class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, heads=2):
        super().__init__()
        self.attn = nn.MultiheadAttention(embed_dim, heads, batch_first=True)
        self.norm1 = nn.LayerNorm(embed_dim)
        self.ff = nn.Sequential(
            nn.Linear(embed_dim, 4 * embed_dim),
            nn.ReLU(),
            nn.Linear(4 * embed_dim, embed_dim)
        )
        self.norm2 = nn.LayerNorm(embed_dim)
        self.dropout = nn.Dropout(0.1)

    def forward(self, x):
        B, T, _ = x.size()
        attn_mask = generate_causal_mask(B, T, x.device, self.attn.num_heads)
        attn_out, _ = self.attn(x, x, x, attn_mask=attn_mask)
        x = self.norm1(x + self.dropout(attn_out))
        ff_out = self.ff(x)
        return self.norm2(x + self.dropout(ff_out))

class MiniGPT(nn.Module):
    def __init__(self, vocab_size, embed_dim=64, context_size=4, n_layers=2):
        super().__init__()
        self.token_embed = nn.Embedding(vocab_size, embed_dim)
        self.pos_embed = nn.Embedding(context_size, embed_dim)
        self.blocks = nn.Sequential(*[TransformerBlock(embed_dim) for _ in range(n_layers)])
        self.ln_f = nn.LayerNorm(embed_dim)
        self.head = nn.Linear(embed_dim, vocab_size)
        self.context_size = context_size

    def forward(self, x):
        B, T = x.shape
        tok = self.token_embed(x)
        pos = self.pos_embed(torch.arange(T, device=x.device).unsqueeze(0).expand(B, T))
        x = tok + pos
        x = self.blocks(x)
        x = self.ln_f(x)
        return self.head(x[:, -1])  # use final token output

# ✅ Stage 5: Training Setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = MiniGPT(len(vocab)).to(device)
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# ✅ Stage 6: Training Loop
for epoch in range(10):
    total_loss = 0
    for ctx, tgt in loader:
        ctx, tgt = ctx.to(device), tgt.to(device)
        logits = model(ctx)
        loss = loss_fn(logits, tgt)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"✅ Epoch {epoch + 1}, Loss: {total_loss:.4f}")

# ✅ Stage 7: Text Generation
def generate(model, seed_text, length=20):
    model.eval()
    ctx = seed_text.lower().split()
    for _ in range(length):
        if len(ctx) < context_size:
            ctx = ['the'] * (context_size - len(ctx)) + ctx
        input_ids = torch.tensor([word_to_ix.get(w, 0) for w in ctx[-context_size:]]).unsqueeze(0).to(device)
        with torch.no_grad():
            out = model(input_ids)
            pred_id = torch.argmax(out, dim=1).item()
            next_word = ix_to_word[pred_id]
            ctx.append(next_word)
    return ' '.join(ctx)

# 🎯 Generate Sample Text
print("\n📝 Generated:")
print(generate(model, "the court finds"))

# ✅ Stage 8: Perplexity Evaluation
def perplexity(model, loader):
    model.eval()
    loss_total, count = 0, 0
    with torch.no_grad():
        for ctx, tgt in loader:
            ctx, tgt = ctx.to(device), tgt.to(device)
            logits = model(ctx)
            loss = loss_fn(logits, tgt)
            loss_total += loss.item() * ctx.size(0)
            count += ctx.size(0)
    return math.exp(loss_total / count)

print(f"\n📊 Perplexity: {perplexity(model, loader):.4f}")

# ✅ Stage 9: Sample Test Messages
test_prompts = [
    "the court finds",
    "the judge concluded",
    "in light of the",
    "the evidence shows",
    "we therefore hold"
]

print("\n🧪 Sample Test Predictions:")
for prompt in test_prompts:
    generated = generate(model, prompt, length=12)
    print(f"\n🔹 Seed: '{prompt}'\n➡️ {generated}")


```

```python
# Save model and vocab
torch.save(model.state_dict(), "minigpt_transformer.pt")
import pickle
with open("vocab.pkl", "wb") as f:
    pickle.dump((word_to_ix, ix_to_word), f)
print(" Model and vocab saved.")


 #Reload and Predict Without Retraining
#To reuse the model in a new session or script:

# 🧠 Reload model + vocab
import torch, pickle
import torch.nn as nn

# Load vocab
with open("vocab.pkl", "rb") as f:
    word_to_ix, ix_to_word = pickle.load(f)
vocab_size = len(word_to_ix)

# Define model class again (must match saved version)
class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, heads=2):
        super().__init__()
        self.attn = nn.MultiheadAttention(embed_dim, heads, batch_first=True)
        self.norm1 = nn.LayerNorm(embed_dim)
        self.ff = nn.Sequential(
            nn.Linear(embed_dim, 4 * embed_dim),
            nn.ReLU(),
            nn.Linear(4 * embed_dim, embed_dim)
        )
        self.norm2 = nn.LayerNorm(embed_dim)
        self.dropout = nn.Dropout(0.1)

    def forward(self, x):
        B, T, _ = x.size()
        mask = torch.tril(torch.ones(T, T, device=x.device)) == 0
        attn_mask = mask.unsqueeze(0).expand(B * self.attn.num_heads, T, T)
        attn_out, _ = self.attn(x, x, x, attn_mask=attn_mask)
        x = self.norm1(x + self.dropout(attn_out))
        ff_out = self.ff(x)
        return self.norm2(x + self.dropout(ff_out))

class MiniGPT(nn.Module):
    def __init__(self, vocab_size, embed_dim=64, context_size=4, n_layers=2):
        super().__init__()
        self.token_embed = nn.Embedding(vocab_size, embed_dim)
        self.pos_embed = nn.Embedding(context_size, embed_dim)
        self.blocks = nn.Sequential(*[TransformerBlock(embed_dim) for _ in range(n_layers)])
        self.ln_f = nn.LayerNorm(embed_dim)
        self.head = nn.Linear(embed_dim, vocab_size)
        self.context_size = context_size

    def forward(self, x):
        B, T = x.shape
        tok = self.token_embed(x)
        pos = self.pos_embed(torch.arange(T, device=x.device).unsqueeze(0).expand(B, T))
        x = tok + pos
        x = self.blocks(x)
        x = self.ln_f(x)
        return self.head(x[:, -1])

# Init and load weights
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = MiniGPT(vocab_size).to(device)
model.load_state_dict(torch.load("minigpt_transformer.pt", map_location=device))
model.eval()
print("✅ Model loaded.")

# ✨ Prediction without retraining
def generate(seed_text, length=20):
    ctx = seed_text.lower().split()
    for _ in range(length):
        if len(ctx) < 4:
            ctx = ['the'] * (4 - len(ctx)) + ctx
        ids = torch.tensor([word_to_ix.get(w, 0) for w in ctx[-4:]]).unsqueeze(0).to(device)
        with torch.no_grad():
            out = model(ids)
            next_id = torch.argmax(out, dim=1).item()
            ctx.append(ix_to_word[next_id])
    return ' '.join(ctx)

# 🧪 Try it
print("\n🧪 Inference from saved model:")
print(generate("the court finds", 15))

```
